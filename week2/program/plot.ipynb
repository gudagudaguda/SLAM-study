```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def ratio(n):
    return (2/3*(n**3) + 1/2*(n**2) - 7/6*n) / (n**3)

n_values = np.arange(1, 51)

# Compute the ratio for each value of n
ratios = ratio(n_values)

# Create a plot
plt.plot(n_values, ratios)

# Add labels and title
plt.xlabel('n')
plt.ylabel('Ratio')
plt.title('Ratio of 2/3*n^3+1/2*n^2-7/6*n to n^3')

# Show the plot
plt.show()
```
